# KMeans Cluster

## Imports

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import metrics as mt
from sklearn import cluster as ct
import numpy as np
```

## Load datasets

```python
x = pd.read_csv('./X_dataset.csv')
```

## Model Cluster

```python
n_cluster = np.arange(2,11)
ss_list = []
for c in n_cluster:
    # define
    k_means = ct.KMeans(n_clusters= c)

    # fit
    labels = k_means.fit_predict(x)

    # performance (test)
    ss_avg = mt.silhouette_score(x, labels)
    ss_list.append(ss_avg)
```

```python
# Elbow method

plt.plot(n_cluster, ss_list, marker = 'o')
plt.xlabel('numero de cluster k')
plt.ylabel('silhouete avarage score')
```

```python
k = 7
# define
k_means = ct.KMeans(n_clusters= k)

# fit
labels = k_means.fit_predict(x)

# performance (test)
ss_avg = mt.silhouette_score(x, labels)

print(f'para {k} cluster, A avg ss é de: {ss_avg:.3f}')
```
